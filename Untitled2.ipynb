```python
class Stack:
    class Node:
        def __init__(self, val, next=None):
            self.val = val
            self.next  = next
    
    def __init__(self):
        self.top = None

    def push(self, val):
        self.top = Stack.Node(val, self.top)
        
    def pop(self):
        assert self.top, 'Stack is empty'
        val = self.top.val
        self.top = self.top.next
        return val
    
    def peek(self):
        return self.top.val if self.top else None
    
    def empty(self):
        return self.top == None
    
    def __bool__(self):
        return not self.empty()
    
    def __repr__(self):
        if not self.top:
            return ''
        return '--> ' + ', '.join(str(x) for x in self)
    
    def __iter__(self):
        n = self.top
        while n:
            yield n.val
            n = n.next
            

delim_openers = '{([<'
delim_closers = '})]>'

def check_delimiters(expr):
    """Returns True if and only if `expr` contains only correctly matched delimiters, else returns False."""
    s = Stack()
    for c in expr:
        if c == '{':
            s.push('}')
        elif c == '(':
            s.push(')')
        elif c == '[':
            s.push(']')
        elif c == '<':
            s.push('>')
        elif c in delim_closers:
            if (s.empty() or (c != s.pop())):
                return False
        print(s)
    return s.empty()

check_delimiters('[()][{')
```

```python
def findwater(a):
    n = len(a) #len of array
    lmax = 0 #initialise left maximum value
    rmax = 0 #initilise right max value 
    result = 0
    l= 0  # index
    h = n-1
    while l<=h:
        if a[l] < a[h]:
            if a[l] > lmax:
                lmax = a[l]
            result = result + (lmax -a[l])
            l = l+1
        else:
            if a[h] > rmax:
                rmax = a[h]
            result = result + (rmax - a[h])
            h = h-1
    return result

findwater([0,1,0,2,1,0,1,3,2,1,2,1])
            
        
```

```python
def platform(a,d):
    n = len(a)
    a.sort()
    d.sort()
    result = 1
    plat = 1
    i = 1
    j = 0
    while(i<n and j<n):
        if a[i] < d[j]:
            plat = plat + 1
            i = i+1
            if plat > result:
                result = plat 
        else:
            plat = plat -1
            j = j+1
    print('The minimum number of platform is',result) 

platform([900, 940, 950, 1100, 1500, 1800],[910, 1200, 1120, 1130, 1900, 2000])
```

```python
def rev(a):
    n = len(a)
    l =[]
    for i in a:
        l.append(i)
    for j in range(n//2):
        l[j],l[n-j-1] = l[n-j-1],l[j]
    return ''.join(l)

rev('ayush')
```
